## Conditional Statements:
    - To check whether the given condition is either True or False
    - if => single condition
    - if-else => two cases
    - elif => n number of cases and based on preference
    - nested if => n number of cases

## if
    Syntax:
        if condition:
            //stmnts
        () => optional

````python
k = int(input("Enter a number: "))
if k<30:
    print(f"Entered number {k} is less than 30")
````

## elif
    Syntax:
     if condition-1:
         //stmnt-1
     elif condition-2:
         //stmnt-2
         |       |
     elif condition-n:
         //stmnt-n
     else:
         //stmnt-n-1

````python
n = int(input())
if n<=40:
    print(f"Entered number {n} is Less than 40")
elif n%2==0 and n%4==0:
    print(f"Entered number is even and divisible by 4")
elif n%2==0:
    print("Even")
else:
    print("Odd")
````

## Loops:
    - To iterate the n number of statements for n number of times untill the condition is false
    - for => for known ranges
    - while => for known and unknown ranges, infinity loops

## for:
    syntax:
        for iterationvariable_name in range(start,stop,step):
            //statements
                        or
        for iterationvariable_name in iterator:
            //statements
        
        Iterators => list,tuple,set,string and dictionary

## range:
    range(stop):
        - By default it starts with 0 value
        - It ends with stop-1 value
        - By default it increments with +1 value
        Ex:
            range(4)
            0 1 2 3
    range(start,stop):
        - It starts exactly with start value
        - It ends with stop-1 value
        - By default it increments with +1 value
        Ex:
            range(4,9)
            4 5 6 7 8
    range(start,stop,step):
        - It starts exactly with start value
        - It ends with stop-1 value
        - It increments/decrements with step value
        Ex:
            range(1,10,3)
            1 4 7 

````python
n = int(input("Enter a range: "))
for i in range(n):
    print(i)
````

````python
n = int(input("Enter a range: "))
for i in range(n):
    print(i,end=" ")
````

````python
n = int(input("Enter a range: "))
for i in range(n):
    print(i+1,end=" ")
````

````python
n = int(input())
for i in range(n):
    if i%2==0:
        print("0",end=" ")
    else:
        print("-1",end=" ")
````

````python
t = int(input("Enter start range: "))
m = int(input("Enter end range: "))
for i in range(t,m):
    print(i,end=" ")
````

## While:
    Syntax:
        variable_initialization
        while condition:
            //stmnts
            incr/decr

````python
w = int(input("Enter a range: "))
i = 0
while i!=w:
    print(i,end=" ")
    i+=1
````

````python
m = int(input())
while m!=0:
    m-=1
    print(m,end=" ")
````
